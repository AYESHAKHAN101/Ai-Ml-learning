# Dummies variable using pandas

```python
df=pd.read_csv('homepricetown.csv')
df
```

# Reverte town


```python
dumiesdf= pd.get_dummies(df.town)
#dumiesdf[['monroe township']]= dumiesdf['monroe township'].replace({False:0 , True :1})
dumiesdf=dumiesdf.astype(int)
dumiesdf
```

```python
#dr0p
```

```python
#concatinate
merge = pd.concat([df, dumiesdf], axis = 'columns')
merge
```

```python
final = merge.drop(['town', 'west windsor'], axis = 'columns')
final
```

```python
y = final.price
```

```python
#x=area,town--->price prediction
x=final.drop('price', axis = 'columns')

#if i want to goback to the vinterser 2 zeros =wintswe
#0,0 =wintswe
#0,1 = robinsville 
#1,0 =monroe 
#sqfeet,area,town 
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict(x)
```

```python
#Accuracy
model.score(x,y)
```
